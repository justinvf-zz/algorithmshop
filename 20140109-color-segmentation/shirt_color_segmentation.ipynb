## What color is my shirt?

This example goes through figuring out what color different regions of an image are.

Many parts of this code from @stefenvdwalt, specifically his pydata2012 presentation.

```python
FILE_PATH = 'anders_shirt_color.jpg'

# Higher number => less detail. Really just the pyramid level.
DETAIL = 3
CLUSTERS = 5
```

```python
from skimage import io, color, exposure
from skimage.transform import pyramid_gaussian
from sklearn import cluster, preprocessing
import numpy as np

img = io.imread(FILE_PATH)

pyramid = tuple(pyramid_gaussian(img, downscale=2))
img = pyramid[DETAIL]

# We are going to cluster in a different space than rgb.
(to_other_space, to_rgb) = (color.rgb2lab, color.lab2rgb)
space_2_dimensions = [1, 2]
brightness_dimension = 0
(space_min, space_max) = (-100, 100)

# Uncomment to instead use hsv
#(to_other_space, to_rgb) = (color.rgb2hsv, color.hsv2rgb)
#space_2_dimensions = [0, 1]
#brightness_dimension = 2
#(space_min, space_max) = (0,1)

# convert to LAB colorspace
space_2_img = to_other_space(img)

# Drop the L componenet, just want to cluster on the a and b components
clusterable_points = space_2_img[:,:,space_2_dimensions]

X = clusterable_points.reshape(-1, 2)
kmeans = cluster.MiniBatchKMeans(n_clusters=CLUSTERS)
y = kmeans.fit(X).labels_

labels = y.reshape(img.shape[:2])
```

```python
# Originally by @stefanvdwalt from auto-clustering suggestion by  Matt Terry
# Hacked up for a blog post by @justinvf

# Make some images with the different cluster highlighted in the original image.
from skimage import morphology


def get_image_and_mask(cluster):
    mask = labels == cluster
    # some morphology stuff really cleans up the output picture
    mask = morphology.binary_closing(mask, morphology.diamond(1))
    mask = morphology.binary_opening(mask, morphology.diamond(1))
    # Binary opening/closing outputs non binary values
    mask = mask.astype(np.bool)
    img_cluster = space_2_img.copy()

    # Darken the points that are not in the cluster by down
    # scaling the L component and knocking out the color
    # components
    img_cluster[..., brightness_dimension][~mask] *= .5
    # knock out the non brightness components
    for d in space_2_dimensions:
        img_cluster[:,:, d][~mask] = 0

    to_return = img.copy()
    to_return[~mask] = to_rgb(img_cluster)[~mask]
    return (to_return, mask)


(img_height, img_width) = img.shape[:2]

# This is going to be our sprited blog post image
all_images = np.zeros((img_height * (CLUSTERS + 1), img_width, 3))

# Add original image at top
all_images[0:img_height, :, :] = img

# Make a image with all the clusters
for i in range(1, CLUSTERS + 1):
    (as_rgb, _) = get_image_and_mask(i - 1)
    all_images[img_height * i:img_height * (i + 1), :, :] = as_rgb

# All the shirt pics in one image for spriting
io.imsave('all_shirt_sprites.jpg', all_images)
```

```python
# Make the histogram
import matplotlib.pyplot as plt

#(_, a, b) = img_lab.T
a = space_2_img[..., space_2_dimensions[0]]
b = space_2_img[..., space_2_dimensions[1]]

print(a.min(), a.max())

left, right, n_bins = space_min, space_max, 400
bins = np.linspace(left, right, n_bins)

H, x_edges, y_edges = np.histogram2d(x=a.flatten(),
                                     y=b.flatten(),
                                     bins=bins,
                                     normed=True)

# No clue why I do this, but I read it in the matplotlib implementation
# and now everything works. Need to think about this later,
# but sick of the problem for now.
H = H.T

H_bright = exposure.rescale_intensity(H, in_range=(0, 5e-4))

size = 5 * 192 / 293
plt.figure(figsize=(size,size))
ax = plt.subplot(111)
ax.imshow(H_bright,
          extent=[-100, 100, -100, 100],
          cmap=plt.cm.binary,
          origin='lower')

ax.set_xticks([])
ax.set_yticks([])
#plt.xlabel('A')
#plt.ylabel('B')

plt.savefig('histogram.png', bbox_inches='tight')
```

```python
# We want some human names for the colors. Scrape Wikipedia!

from bs4 import BeautifulSoup
from httplib2 import Http
import re

# First lets get a list of named colors
COLOR_LIST = 'http://en.wikipedia.org/wiki/List_of_colors_(compact)'

(r, content) = Http().request(COLOR_LIST)
assert(r['status'] == '200')

wiki_soup = BeautifulSoup(content, 'html5lib')

RGB_REGEX = re.compile(r'rgb\((\d+),(\d+),(\d+)\)')

name_and_rgb = []

for d in wiki_soup.find_all('div'):
    paragraphs = d.find_all('p')
    if len(paragraphs) != 2:
        continue
    (p_0, p_1) = paragraphs
    rgb = RGB_REGEX.search(p_0.get('style'))
    if rgb and p_1.text:
        rgb = list(map(int, rgb.groups()))
        name_and_rgb.append((p_1.text, rgb))

print('{} named colors.\nExamples:'.format(len(name_and_rgb)))
                               
name_and_rgb[::len(name_and_rgb) // 15]
```

```python
# We want to find the closest named color for each pixel in a cluster.
(color_names, rgb) = zip(*name_and_rgb)

rgb_array = np.array(rgb).astype(np.uint8)

from scipy.spatial import cKDTree

color_tree = cKDTree(rgb_array)

# for debugging, shows the image rounded to the named colors
rounded_image = None

def top_n_colors(pixels, num_results=10, debug=False):
    global rounded_image
    if pixels.max() < 1:
        # We are in [0-1] color range
        pixels = pixels * 256
    (_, colors) = color_tree.query(pixels)
    if debug:
        rounded_image = rgb_array.take(colors, axis=0)
    
    counts = np.bincount(colors.flatten())
    top_k = np.argsort(counts)[-num_results:][::-1]
    return [(color_names[i], counts[i]) for i in top_k]
```

```python
top_n_colors(img, debug=True)
```

```python
# Here's the rounded image. Looks pretty close to the original.
plt.imshow(rounded_image)
```

```python
SIZE = 8
plt.figure(figsize=(SIZE, CLUSTERS*SIZE))
NUM_COLORS = 5

def draw_for_cluster(cluster, n_rows, n_cols, start_index):
    """Plot the information about the cluster, and return some data."""

    (as_rgb, mask) = get_image_and_mask(cluster)
    
    color_counts = top_n_colors(as_rgb[mask], num_results=NUM_COLORS)

    plt.subplot(n_rows, n_cols, start_index)
    plt.imshow(as_rgb)
    plt.axis('off')

    # Draw the histogram
    (color_labels, counts) = zip(*color_counts)
    counts = np.array(counts)
    counts = counts / np.linalg.norm(counts)

    plt.subplot(n_rows, n_cols, start_index + 1)
    plt.axis('on')
    plt.bar(np.arange(NUM_COLORS), counts)
    plt.xticks(np.arange(NUM_COLORS), color_labels, rotation=20)
    plt.tick_params(labelsize=12)

d3_json = []

for i in range(CLUSTERS):
    draw_for_cluster(i, CLUSTERS, 2, 2 * i + 1)

plt.savefig('cluster_breakdowns.png')
```

```python
import json

# Get some data for d3 blog output.
NUM_COLORS = 5

cluster_stats = []

used_colors = set()

for i in range(CLUSTERS):
    (as_rgb, mask) = get_image_and_mask(i)
    color_counts = top_n_colors(as_rgb[mask], num_results=NUM_COLORS)
    used_colors.update(x[0] for x in color_counts)
    cluster_stats.append({'cluster': i,
                          'cluster_center': list(map(float, kmeans.cluster_centers_[i])),
                          'counts': [{"count": int(v), "color": k}
                                     for (k,v) in color_counts],
                          'size': int(np.sum(mask))})

with open('cluster_data.json', 'wt') as f:
    used_color_dict = dict((k,v) for (k,v) in name_and_rgb if k in used_colors)
    json.dump({'color_stats': used_color_dict,
               'cluster_stats': cluster_stats}, f)
```
